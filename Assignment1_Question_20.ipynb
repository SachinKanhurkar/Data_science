```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
data=pd.read_csv("Cars.csv")
```

```python
data.describe()
```

```python
data
```

```python
n =max(data.shape)
```

```python
n
```

```python
count = 0
for i in range(0, n):
    if data.iloc[i, 1]>38:
        count +=1
print(count)
prob= count/n
print(prob)   
```

```python
data.iloc[0, 1]
```

```python
count = 0
for i in range(0, n):
    if data.iloc[i, 1]<40:
        count +=1
print(count)
prob= count/n
print(prob)   
```

```python
count = 0
for i in range(0, n):
    if data.iloc[i, 1]>20 and data.iloc[i, 1]< 50 :
        count +=1
print(count)
prob= count/n
print(prob)  
```

```python
for v in range(0,len(data.columns)):
    x = data[data.columns[v]]
    #plt.hist(x)
    sns.displot(x, label='{}'.format(data.columns[v]))
    plt.xlabel('{}'.format(data.columns[v]))
    plt.ylabel('Frequency')    
    plt.title('Histogram of {}'.format(data.columns[v])) 
    plt.legend()
    plt.show()
```

```python
for v in range(0,len(data.columns)):
    x = data[data.columns[v]] 
    sns.set(rc={"figure.figsize": (8, 4)}); np.random.seed(0)
    #plt.hist(x)
    sns.displot(x, label='{}'.format(data.columns[v]),  kde=True)
    plt.xlabel('{}'.format(data.columns[v]))
    plt.ylabel('Frequency')    
    plt.title('Histogram of {}'.format(data.columns[v]))       
    plt.grid(True)
    plt.legend()   
    plt.show()
```

```python
for i in range(0, len(data.columns)):
    a= data.columns[i]
    print("For the {} dataset:-".format(a))
    print("Mean of {}: ".format(a), data[a].mean())
    print("Mode of {} ".format(a), data[a].mode())
    print("Median of {} ".format(a), data[a].median())
    print("Variance of {} ".format(a), data[a].var())
    print("Standard deviation of {} ".format(a), data[a].std())
    ran = max(data[a]) - min(data[a])
    print("Range of {} ".format(a), ran)
    print("\n")
```
